## 3. Data Cleaning

```python
learning_df.rename(columns={
    'Course_Completion_Rate(%)':'Course_Completion_Rate',
    'Satisfaction_Score(1-5)':'Satisfaction_Score'
}, inplace=True)
```

```python
learning_df['Enrollment_Date'] = pd.to_datetime(learning_df['Enrollment_Date'])
```

```python
learning_df.info()
```

```python
learning_df.groupby('Field_of_Study')['Course_Completion_Rate'].mean().round(2).sort_values(ascending=False)
```

```python
learning_df.groupby('Field_of_Study')['Course_Completion_Rate'].mean().round(2).sort_values(ascending=False).reset_index()
```

```python
def categorize_job(title):
    if 'Scientist' in title:
        return 'Data Science'
    elif 'AI' in title or 'ML' in title or 'Machine Learning' in title:
        return 'AI/ML Engineer'
    elif 'Data Engineer' in title or 'ETL' in title:
        return 'Data Engineer'
    elif 'Analyst' in title:
        return 'Data Analyst'
    else:
        return 'Other'
```

```python
salary_df['Field_of_Job'] = salary_df['job_title'].apply(categorize_job)
```

```python
salary_df.columns
```

```python
# Count number of jobs in each category
```

```python
job_count = salary_df['Field_of_Job'].value_counts().sort_values(ascending=False).reset_index()
job_count.columns = ['Field_of_Job','job_count']
job_count
```

```python
# Average satisfaction score per field
```

```python
learning_df.columns
```

```python
learning_df.groupby('Field_of_Study')['Satisfaction_Score'].mean().round(2).sort_values(ascending=False).reset_index()
```

```python
#Remarks - Satisfaction scores are relatively consistent across fields, with only minor variation. Arts shows the highest satisfaction while Business is the lowest, indicating that field of study alone may not strongly influence satisfaction levels.
```

```python
# Grouping by study hour range to find optimal learning time
```

```python
learning_df.groupby('Daily_Learning_Hours')['Course_Completion_Rate'].mean().round(2).sort_values(ascending=False).reset_index()
```

```python
learning_df['hour_range']=pd.cut(
    learning_df['Daily_Learning_Hours'],
    bins=[0,1,2,3,4,5],
    labels=['0-1','1-2','2-3','3-4','4-5']
)
```

```python
learning_df.groupby('hour_range',observed=False)['Course_Completion_Rate'].mean().round(2).sort_values(ascending=False).reset_index()
```

```python
#Remarks - Completion rates peak in the 2–3 hour range, and do not consistently increase with more study time, indicating diminishing returns beyond optimal study duration.
```

## 7. Final Conclusion

This analysis highlights a clear imbalance between learning outcomes and job market demand. While Data Science and related roles show the highest demand in the job market, their course completion rates are comparatively lower, indicating a gap between student interest and successful skill acquisition.

In contrast, foundational fields such as Computer Science and Engineering demonstrate both high completion rates and strong alignment with technical job opportunities, making them more reliable pathways for career entry into high-demand roles.

Additionally, fields like Arts and Business, despite moderate completion rates, show weaker direct alignment with specific high-growth job categories, suggesting potential career shifts or broader competition in the job market.

Overall, the findings suggest that job demand alone is not a sufficient indicator of learning success. Instead, a combination of strong foundational skills, practical exposure, and structured learning support is essential—especially in complex and high-demand fields like Data Science.
